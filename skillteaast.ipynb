#1. 動的計画法用の問題設定
📂 Project 1: 大規模状態遷移における経路総数の計算
##【問題設定 (Problem Statement)】
2つの状態（状態A、状態B）の間を移動するシステムを考える。初期状態（1日目）において、状態Aへの遷移パターンが6通り、状態Bへの遷移パターンが5通り存在する。
2日目以降、各状態は前日の状態に応じて以下のルールで遷移する。

翌日の状態Aのパターン数 = (前日のA × 2) + (前日のB × 4)

翌日の状態Bのパターン数 = (前日のA × 3) + (前日のB × 2)

このとき、指定されたステップ数（例: 1461ステップ後）における、取り得るすべての遷移パターンの総数を求めよ。ただし、ステップ数が大きくなると組み合わせ数が爆発的に増加するため、計算効率を考慮すること。

```python
def simulate_state_transition(steps=1461):
    """
    動的計画法（DP）を用いた状態遷移の組み合わせ計算
    2つの状態(A, B)間の遷移ルールに基づき、指定ステップ後の取り得る全ルート数を計算する。
    """
    # 初期状態のパターン数
    state_a_patterns = 6
    state_b_patterns = 5

    # DPによる状態遷移の計算 (O(N))
    for _ in range(2, steps + 1):
        # 次のステップへの遷移ルール（固定の係数を用いた線形結合）
        next_a = state_a_patterns * 2 + state_b_patterns * 4
        next_b = state_a_patterns * 3 + state_b_patterns * 2

        state_a_patterns, state_b_patterns = next_a, next_b

    total_patterns = state_a_patterns + state_b_patterns

    print(f"【実行結果】")
    print(f"計算ステップ数: {steps}")
    print(f"最終的な総パターン数（概算）: 約 {float(str(total_patterns)[:5])} × 10^{len(str(total_patterns))-1}")

# 実行
simulate_state_transition()
```

#2. Two-pointers（容量最適化）用の問題設定
📂 Project 2: 貪欲法とTwo-pointersによる容量詰め込み最適化（ビンパッキング問題）
##【問題設定 (Problem Statement)】
容量が最大 1.0 であるコンテナが大量に存在する。ここに、0.0 から 1.0 までのランダムな量の水（または荷物のサイズ）が入ったデータが 1,000,000 件与えられる。
複数のコンテナの水を1つのコンテナにまとめることで、最終的に「水が少しでも入っているコンテナの総数」をできるだけ少なくしたい（効率よく集約したい）。
100万件という大規模なデータセットに対して、実行時間1秒未満で効率的な近似解を出力するアルゴリズムを構築せよ。

```python
import numpy as np
import time

def optimize_capacity_with_two_pointers(n_samples=1000000):
    """
    Two-pointers（尺取り法）と貪欲法を用いた容量最適化アルゴリズムの実装検証
    計算量: O(N log N)
    """
    # 1. データの準備
    np.random.seed(42)
    # 処理速度向上のため、Numpy配列からPythonの標準リストへ変換
    capacities = np.random.rand(n_samples).tolist()

    print(f"【実行条件】")
    print(f"データサイズ: {n_samples:,} 件 (0.0 ~ 1.0のランダム値)")

    start_time = time.time()

    # 2. アルゴリズム本体
    # 降順ソート (計算量: O(N log N))
    capacities.sort(reverse=True)

    # 両端にポインタを配置
    left = 0
    right = len(capacities) - 1

    # ポインタが交差するまで貪欲に容量を統合 (計算量: O(N))
    while left < right:
        while left < right and capacities[left] + capacities[right] <= 1.0:
            capacities[left] += capacities[right]
            capacities[right] = 0.0
            right -= 1

        left += 1

    # 3. 結果の集計
    active_count = sum(1 for c in capacities if c > 0)

    end_time = time.time()

    print(f"【実行結果】")
    print(f"最適化後の有効データ数: {active_count:,} 件")
    print(f"実行時間: {end_time - start_time:.4f} 秒")

# 実行
if __name__ == "__main__":
    optimize_capacity_with_two_pointers()
```
